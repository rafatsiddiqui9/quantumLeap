```python
# Extract URLs, 'days ago' information and transcripts
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import re
import time

# Set up Selenium webdriver
driver = webdriver.Chrome()  # Make sure you have chromedriver installed

video_data = []
for item in video_items:
    # Extract the video URL
    link = item.find('a', id='thumbnail')
    url = "https://www.youtube.com" + link['href'] if link else None
    
    # Extract 'days ago' information
    metadata_items = item.find_all('span', class_='inline-metadata-item style-scope ytd-video-meta-block')
    days_ago = None
    for metadata in metadata_items:
        if 'ago' in metadata.get_text():
            days_ago = metadata.get_text(strip=True)
            break
    
    if url and days_ago:
        try:
            # Navigate to video page
            driver.get(url)
            
            # Click on "..." button to open menu
            more_button = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "button.ytp-button.ytp-settings-button"))
            )
            more_button.click()
            
            # Click on "Show transcript" option
            transcript_button = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, "//div[contains(text(), 'Show transcript')]"))
            )
            transcript_button.click()
            
            # Wait for transcript to load and get text
            transcript_elements = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div.segment-text"))
            )
            
            transcript = ' '.join([elem.text for elem in transcript_elements])
            
        except TimeoutException:
            transcript = None
        except Exception as e:
            print(f"Error extracting transcript for {url}: {str(e)}")
            transcript = None
            
        video_data.append({
            "url": url,
            "days_ago": days_ago, 
            "transcript": transcript
        })
        
        # Add small delay between requests
        time.sleep(2)

# Close browser
driver.quit()

# Create pandas DataFrame
import pandas as pd
df = pd.DataFrame(video_data)
```
